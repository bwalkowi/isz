```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
plt.rcParams["figure.figsize"] = [16, 9]
```

```python
def population_growth(r, generations=20, initial_population=0.5):
    populations = []
    size = initial_population
    for _ in range(generations):
        populations.append(size)
        size = r * size * (1 - size)
    return populations
```

```python
rates = (0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5)
for r in rates:
    plt.plot(population_growth(r), label=r)
    plt.legend()
plt.show()
```

```python
plt.plot(population_growth(3.8))
plt.show()
```

```python
def population_growth2(r, iter1=100, iter2=100):
    trash = population_growth(r, generations=iter1)
    observation = population_growth(r, generations=iter2, initial_population=trash[-1])
    return observation
```

```python
rates = np.linspace(0.5, 4, 1000)

x = []
y = []
for r in rates:
    for val in set(population_growth2(r)):
        x.append(r)
        y.append(val)

plt.scatter(x,y, s =0.05, color='black')
plt.show()
```

```python
rates = np.linspace(3.84, 3.855, 1000)

x = []
y = []
for r in rates:
    for val in set(population_growth2(r)):
        x.append(r)
        y.append(val)

plt.scatter(x,y, s =0.05, color='black')
plt.show()
```

```python
import random

r = 3.99

vals1 = population_growth2(r, iter1=40, iter2=260)
plt.plot(vals1, label='pg')
plt.legend()

vals2 = [random.random() for _ in range(40, 300)]
plt.plot(vals2, label='rand')
plt.legend()

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for vals, color, label in ((vals1, 'r', 'pg'), (vals2, 'b', 'rand')):
    xs = vals[:-2]
    ys = vals[1:-1]
    zs = vals[2:]
    ax.scatter(xs, ys, zs, c=color, label=label)
    ax.legend()
```
